```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import math
from tensorflow.keras.callbacks import LearningRateScheduler, ReduceLROnPlateau

epochs = np.arange(0, 30)
```

```python
def step_decay(epoch):
    initial_lr = 0.001
    drop = 0.5
    epochs_drop = 5
    return initial_lr * math.pow(
        drop,
        math.floor((1 + epoch) / epochs_drop)
    )

step_lr = [step_decay(i) for i in epochs]
```

```python
def exponential_decay(epoch):
    initial_lr = 0.001
    k = 0.1
    return initial_lr * math.exp(-k * epoch)

exp_lr = [exponential_decay(i) for i in epochs]
```

```python
plateau_lr = []
lr = 0.001

for epoch in epochs:
    if epoch in [7, 14, 21, 26]:
        lr *= 0.5
    plateau_lr.append(lr)
```

```python
plt.plot(epochs, step_lr, marker='o', label='Step Decay')
plt.plot(epochs, exp_lr, marker='s', label='Exponential Decay')
plt.plot(epochs, plateau_lr, marker='^', label='Reduce on Plateau')

plt.title('Learning Rate Scheduling Techniques')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.legend()
plt.grid(True)
plt.show()
```
